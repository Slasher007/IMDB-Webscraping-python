# Recup√©ration d'une liste Ips pour eviter d'etre banni par le serveur : Ip Rotating 

```python
#Proxy_Nova
def get_proxies_proxy_nova ():

    #url_proxy = 'https://www.proxynova.com/proxy-server-list/country-fr/'

    #Anonymous only
    url_proxy ="https://www.proxynova.com/proxy-server-list/anonymous-proxies/"

    #Elite only
    #url_proxy = "https://www.proxynova.com/proxy-server-list/elite-proxies/"

    re = requests.get(url_proxy)
    soup = bs(re.content,'html.parser')
    allproxy = soup.find_all("tr")

    proxies_list = []
    for i in range (1,36):

        ip = allproxy[i].find("script").text.replace("document.write(","").replace(")","").replace("\'","").replace(";","")
        try :
            port = allproxy[i].find('td').next_sibling.next_sibling.text.replace(" ","").replace("\n","")
            anonymity = allproxy[i].find_all('span')[2].text
            http_proxy = "http://"+ip+":"+port
            #Check anonymity
            #if(anonymity == 'Elite' or anonymity == 'Anonymous'):
            proxies_list.append(http_proxy)
        except :
            pass
  
    return proxies_list
```

# Fonction pour obtenir la soup avec url: obtenir_soup_url (url,proxy_list) 

```python
def obtenir_soup_url (url_,proxy_list):
    
    if proxy_list != None:
        headers = random.choice(headers_list)
        #r = requests.Session()
        #r.headers = headers
        good_proxy = False

        while (good_proxy == False):
            p = random.choice(proxy_list)
            try : 
                r = requests.get(url=url_,proxies = {"http":p},headers=headers,stream=True,timeout=1)
                print("\nTarget : ",url_)
                try :
                    print("Server:", r.raw._connection.sock.getsockname())
                    print("User-Agent Sent:%s"%(headers['User-Agent']))
                    soup = bs(r.content,'html.parser')
                    good_proxy = True
                except:
                    print("No data received")
                    pass
                    good_proxy = False
                print("Request ip : ",p)
            except:
                good_proxy = False
                #proxy_list.remove(p)
    else:
          r = requests.get(url_)
          soup = bs(r.content,"html.parser")

    return soup
```

# Fonction pour recup√®rer urls de la liste des films en ligne : obtenir_urlfilms_avec_url(url)

```python
def obtenir_urlfilms_avec_url(url):

    soup = obtenir_soup_url(url,None)

    href_films = []
    liste_films = soup.findAll('div',attrs={'class':'lister-item mode-advanced'})
    for l in liste_films:
        lien = l.h3.a.get("href")
        href_films.append(url_page_accueil+lien.replace("?ref_=adv_li_tt",""))

    return href_films
```

# Fontion qui r√©cup√©re les details pour chaque film : def obtenir_details_film(soup)

```python
def obtenir_details_film(soup):
  
    details_films = {}
    details_films["nom"] = soup.find('title').text.replace(" - IMDb","")
    details_films["note"] = float(soup.find('span',{'class':'AggregateRatingButton__RatingScore-sc-1ll29m0-1 iTLWoV'}).text)

    details_films["nombre_de_votes"] = soup.find("div",{"class":"AggregateRatingButton__TotalRatingAmount-sc-1ll29m0-3 jkCVKJ"}).text
    if 'K' in details_films["nombre_de_votes"] :
        details_films["nombre_de_votes"] = details_films["nombre_de_votes"].replace("K","")
        details_films["nombre_de_votes"] = float(details_films["nombre_de_votes"]) * 1000
    elif 'M' in details_films["nombre_de_votes"] :
        details_films["nombre_de_votes"] = details_films["nombre_de_votes"].replace("M","")
        details_films["nombre_de_votes"] = float(details_films["nombre_de_votes"]) * 1000000
    else:
        details_films["nombre_de_votes"] = float(details_films["nombre_de_votes"])

    details_films["nombre_de_reviews"] = soup.find('span',{'class':'score'}).text
    if 'K' in details_films["nombre_de_reviews"] :
        details_films["nombre_de_reviews"] = details_films["nombre_de_reviews"].replace("K","")
        details_films["nombre_de_reviews"] = float(details_films["nombre_de_reviews"]) * 1000
    elif 'M' in details_films["nombre_de_reviews"] :
        details_films["nombre_de_reviews"] = details_films["nombre_de_reviews"].replace("M","")
        details_films["nombre_de_reviews"] = float(details_films["nombre_de_reviews"]) * 1000000
    else:
        details_films["nombre_de_reviews"] = float(details_films["nombre_de_reviews"])

    details_films["annee_sortie"] = int(soup.find('span',{"class":"TitleBlockMetaData__ListItemText-sc-12ein40-2 jedhex"}).text)
    details_films['directeur'] = soup.find('a',{'class':"ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"}).text
    details_films["date_sortie"] = soup.find('li',{"data-testid":"title-details-releasedate"}).text.replace("Release date","")
    details_films["pays_de_sortie"] =  soup.find('li',{"data-testid":"title-details-origin"}).a.text
    release_country = details_films["pays_de_sortie"]
    details_films["date_sortie"] = details_films["date_sortie"].replace(release_country,"").replace("()","")
    genres = soup.find('li',{"data-testid":"storyline-genres"}).find_all("a")

    genres = soup.find('li',{"data-testid":"storyline-genres"}).find_all("a")
    liste_genres = []
    for g in genres:
        liste_genres.append(g.text)

    liste_genres
    details_films["genre"] = liste_genres
    try:
        details_films["lieu_de_tournage"] = soup.find("li",{"data-testid":"title-details-filminglocations"}).find("a",{"class":"ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"}).text
    except:
        details_films["lieu_de_tournage"] = None
    
    writters = soup.findAll('ul',{"class":"ipc-inline-list ipc-inline-list--show-dividers ipc-inline-list--inline ipc-metadata-list-item__list-content baseAlt"})[1]
    w = writters.find_all('a')

    liste_scen = []
    for e in w:
        liste_scen.append(e.text)

    liste_scen

    details_films["scenaristes"] = liste_scen

    details_films["duree"] = soup.find("li",{"data-testid":"title-techspec_runtime"}).span.next_sibling.text


    try :
        budget = soup.find("li",{"data-testid":"title-boxoffice-budget"}).span.next_sibling.text
        details_films['budget'] = budget
    except:
        details_films['budget'] = None
    
    try :
        gross_world = soup.find("li",{"data-testid":"title-boxoffice-cumulativeworldwidegross"}).span.next_sibling.text
        details_films["revenue_monde"] = gross_world
    except:
        details_films["revenue_monde"] = None

    try :
        gross_usa = soup.find("li",{"data-testid":"title-boxoffice-grossdomestic"}).span.next_sibling.text
        details_films["revenue_usa"] = gross_usa
    except:
        details_films["revenue_usa"] = None

    try :
        gross_week_usa = soup.find("li",{"data-testid":"title-boxoffice-openingweekenddomestic"}).span.next_sibling.span.text
        details_films["revenue_week_usa"] = gross_week_usa
    except:
        details_films["revenue_week_usa"] = None
    
    try :
        details_films["sound_mix"] = soup.find("li",{"data-testid":"title-techspec_soundmix"}).a.text
    except:
        details_films["sound_mix"] = None
    
    try :
        details_films["color"] = soup.find("li",{"data-testid":"title-techspec_color"}).a.text
    except:
        details_films["color"] = None

    try :
        details_films["aspect_ratio"] = soup.find("li",{"data-testid":"title-techspec_aspectratio"}).span.next_sibling.text
    except:
        details_films["aspect_ratio"] = None

    return details_films
```

# fonction pour cr√©er la base de donn√©es de tous les films:




```python
#Fonction Creation de la dataFrame √† partir d'une liste urls:

def creation_df_urls(url_cibles):
    df_films = pd.DataFrame(columns=['nom','genre','scenaristes','directeur','annee_sortie','date_sortie','pays_de_sortie','lieu_de_tournage','duree','note','nombre_de_votes'
           ,'nombre_de_reviews','budget','revenue_monde','revenue_usa','revenue_week_usa','sound_mix','color','aspect_ratio',])
    

    #my_proxies = get_proxies_proxy_nova()
    for url in url_cibles:
        #soup = obtenir_soup_url(url,my_proxies)
        soup = obtenir_soup_url(url,None)
        #try :
        dict_film = obtenir_details_film(soup)
        film = pd.DataFrame.from_dict([dict_film])
        df_films = df_films.append(film,ignore_index=True)
        #except:
        #    print("%s failed ",url)
        #    pass

    return df_films
```
